```python
path="C://t//test//data1"
```

```python
op=open(path+"//sample1.txt","w")
op.write("Hello world\n")
op.write("Welcome\n")
op.close()
```

```python
op=open(path+"//sample1.txt","a")
op.write("Hello world\n")
op.write("Welcome\n")
op.close()
```

```python
ip = open(path+"//sample1.txt","r")
content=ip.read()
ip.close()
print(content)
```

```python
ip = open(path+"//sample1.txt","r")
for line in ip:
    print(line.strip())
ip.close()
```

```python
with open(path+"//sample1.txt","r") as ip:
    for line in ip:
        print(line.strip())
```

```python
with open(path+"//sample1.txt","a+") as ip:
    ip.write("new line \n")
    ip.seek(0)
    content=ip.read()
    print(content)
```

```python
with open(path+"//sample1.txt","a+") as ip:
    ip.write("new line \n")
    ip.seek(0)
    for line in ip:
        print(line)
```
